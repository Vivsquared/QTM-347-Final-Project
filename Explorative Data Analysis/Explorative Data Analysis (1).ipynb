```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
#load the dataset
df = pd.read_csv('dataQTM.csv')
df = df.drop(['id', 'Unnamed: 32'], axis=1)
```

```python
#display basic information about the dataset
print(df.info())
```

```python
#display summary statistics
print(df.describe())
```

```python
#check for missing values
print("Missing Values:")
print(df.isnull().sum())
```

```python
#check the distribution of the dependent variable
df['diagnosis'].value_counts()
```

```python
#generate a plot to show the distribution of diagonosis
sns.countplot(x='diagnosis', data=df)
plt.title('Distribution of Diagnosis')
plt.show()
```

```python
#change B to 0 and M to 1 in diagonsis to find correlation between all the variables
df['diagnosis'] = df['diagnosis'].replace({'B': 0, 'M': 1})
df.corr()
```

```python
#generate a heat map to show the correlation between disgnosis and all the mean variables
df_mean = df[['diagnosis','radius_mean', 'texture_mean', 'perimeter_mean', 'area_mean','smoothness_mean', 'compactness_mean', 'concavity_mean', 'concave points_mean', 'symmetry_mean', 'fractal_dimension_mean']]
correlation_mean = df_mean.corr().loc[:,'diagnosis'].sort_values(ascending=False)

plt.figure(figsize=(12, 8))
sns.heatmap(df_mean.corr(), annot=True, cmap='coolwarm', fmt='.2f', linewidths=0.5)
plt.title("Mean Variables Heatmap")
plt.show()
```

```python
#generate a heat map to show the correlation between disgnosis and all the se variables
df_se = df[['diagnosis','radius_se', 'texture_se', 'perimeter_se', 'area_se','smoothness_se', 'compactness_se', 'concavity_se', 'concave points_se', 'symmetry_se', 'fractal_dimension_se']]
correlation_se = df_se.corr().loc[:,'diagnosis'].sort_values(ascending=False)

plt.figure(figsize=(12, 8))
sns.heatmap(df_se.corr(), annot=True, cmap='coolwarm', fmt='.2f', linewidths=0.5)
plt.title("SE Variables Heatmap")
plt.show()
```

```python
#generate a heat map to show the correlation between disgnosis and all the worst variables
df_worst = df[['diagnosis','radius_worst', 'texture_worst', 'perimeter_worst', 'area_worst','smoothness_worst', 'compactness_worst', 'concavity_worst', 'concave points_worst', 'symmetry_worst', 'fractal_dimension_worst']]
correlation_worst = df_worst.corr().loc[:,'diagnosis'].sort_values(ascending=False)

plt.figure(figsize=(12, 8))
sns.heatmap(df_worst.corr(), annot=True, cmap='coolwarm', fmt='.2f', linewidths=0.5)
plt.title("Worst Variables Heatmap")
plt.show()
```

```python
#generate a violin plot of concave points mean for both 0 and 1 in diagonsis
plt.figure(figsize=(15, 8))
sns.violinplot(x='diagnosis', y='concave points_mean', data=df)
plt.title('Violin Plot of Concave Points Mean by Diagnosis')
plt.show()
```
